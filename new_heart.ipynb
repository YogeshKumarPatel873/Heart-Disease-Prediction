```python
data['num']=data['num'].map({1:0,2:0,3:1,4:1})
```

```python
x=data.drop(columns=["num"])
y=data["num"]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=51)
```

```python
from sklearn.preprocessing import MinMaxScaler

sc=MinMaxScaler()
sc.fit(x_train)
x_train=sc.transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
import lightgbm as lgb
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import StackingClassifier
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
```

```python
lrc=LogisticRegression(penalty='l2', max_iter=100, multi_class='ovr', solver='liblinear')
svc=SVC(kernel="poly",gamma=1.0)
mnb=MultinomialNB()
dtc=DecisionTreeClassifier(max_depth=80)
knc=KNeighborsClassifier(n_neighbors=2)
rfc=RandomForestClassifier(n_estimators=200,random_state=2,criterion="entropy")
abc=AdaBoostClassifier(n_estimators=200,random_state=2)
bc=BaggingClassifier(n_estimators=150,random_state=2)
etc=ExtraTreesClassifier(n_estimators=200,random_state=5)
gbc=GradientBoostingClassifier(n_estimators=500,learning_rate=0.8,random_state=2)
xgb=XGBClassifier(n_estimators=300,random_state=5)
lbgm = lgb.LGBMClassifier(boosting_type="dart",n_estimators=200)
vot=VotingClassifier(estimators=[("KN",knc),("XGB",xgb),("RF",rfc),("DT",dtc)],voting="soft")
stack=StackingClassifier(estimators=[("KN",knc),("XGB",xgb),("RF",rfc)],final_estimator=rfc)
```

```python
clfs={"LRC":lrc,"SVC":svc,"MNB":mnb,"KN":knc,"DT":dtc,"RF":rfc,"AdaBoost":abc,"BGC":bc,"ETC":etc,"GBDT":gbc,"XGB":xgb,"LBGM":lbgm,"Voting":vot,"Stacking":stack}
```

```python
def train_classifier(clf,x_train,y_train,x_test,y_test):
    clf.fit(x_train,y_train)
    score=clf.score(x_test,y_test)
    y_pred=clf.predict(x_test)
    accuracy=accuracy_score(y_test,y_pred)
    precision=precision_score(y_test,y_pred,average='micro')
    return score,precision
```

```python
accuracy_score_lst=[]
precision_score_lst=[]
score_1=[]
for name,clf in clfs.items():
    score,precision=train_classifier(clf, x_train, y_train, x_test, y_test)
    accuracy_score_lst.append(score)
    precision_score_lst.append(precision)
```

```python
performance_df=pd.DataFrame({"Algorithm":clfs.keys(),"Accuracy":accuracy_score_lst,"Precision":precision_score_lst}).sort_values('Accuracy',ascending=False)
```
